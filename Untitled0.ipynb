```python
import pdfplumber, re, pandas as pd, json

pdf_path = '/content/drive/My Drive/Colab Notebooks/Sample_Contract.pdf'

# 1. Extract all pages' text
pages_text = []
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        pages_text.append({'page_num': i+1, 'text': text})

# 2. Define helper functions
IS_CODE_REGEX = r'IS\s*\d{3,5}(?:\s*[:\-]?\s*Part\s*\d+)?(?:\s*[:\-]?\s*\d{4})?'
RISKY_PHRASES = [
    "unless otherwise specified", "subject to engineer", "as directed", "unless approved",
    "to the satisfaction", "where applicable"
]
CRITICAL_TEST_WORDS = ["shall be tested", "must conform", "shall comply", "is mandatory", "must be as per"]
OPTIONAL_TEST_WORDS = ["may be tested", "should be", "can be", "as desired"]

def extract_is_codes(text):
    codes = re.findall(IS_CODE_REGEX, text, re.IGNORECASE)
    return list(set([c.strip() for c in codes if c.strip()]))

def flag_critical_tests(text):
    crit = [phrase for phrase in CRITICAL_TEST_WORDS if phrase in text.lower()]
    opt = [phrase for phrase in OPTIONAL_TEST_WORDS if phrase in text.lower()]
    if crit: return "Critical"
    if opt: return "Optional"
    return ""

def extract_tables(text):
    # Looks for lines like "Table 4.3", "Table 3.2", or "gradation table"
    tables = re.findall(r'(Table\s*\d+(\.\d+)?(?:\s*[:\-]?\s*[^\n]*)?)', text)
    return [t[0] for t in tables] if tables else []

def flag_risky(text):
    found = [phrase for phrase in RISKY_PHRASES if phrase in text.lower()]
    return ", ".join(found) if found else ""

def get_clauses(text):
    # Find clause headings like 4.1.1.5 Testing
    return re.findall(r'\d+\.\d+\.\d+\.\d+\s+[A-Za-z ]+', text)

# 3. Main extraction loop
section_pattern = re.compile(r'(\d+\.\d+\.\d+)\s+([A-Za-z ]+)\n')
materials_data = []

for page in pages_text:
    text = page['text'] or ""
    matches = list(re.finditer(section_pattern, text))
    for idx, match in enumerate(matches):
        section_num = match.group(1)
        material = match.group(2).strip()
        start = match.end()
        end = matches[idx+1].start() if idx+1 < len(matches) else len(text)
        content = text[start:end].strip()

        # IS Codes
        is_codes = extract_is_codes(content)
        # Tests
        tests = []
        for sentence in re.split(r'(?<=[.!?]) +', content):
            if re.search(r'test|testing|conform|comply|shall|must|may', sentence, re.IGNORECASE):
                tag = flag_critical_tests(sentence)
                if tag:
                    tests.append(f"{sentence.strip()} [{tag}]")
        # Mix/gradation info
        tables = extract_tables(content)
        # Risk notes
        risks = flag_risky(content)
        # Clauses
        clauses = get_clauses(content)

        materials_data.append({
            "Material": material,
            "Section": section_num,
            "IS Codes": is_codes,
            "Critical IS Codes": [c for c in is_codes if any(w in content.lower() for w in CRITICAL_TEST_WORDS)],
            "Required Tests": tests,
            "Gradation/Mix Info": tables,
            "Risk Notes": risks,
            "Clauses": clauses,
            "Page Numbers": page['page_num'],
            "Raw Content": content
        })

# 4. Frequency analysis of IS codes
is_code_freq = {}
for mat in materials_data:
    for code in mat["IS Codes"]:
        is_code_freq.setdefault(code, {"count":0, "pages":[]})
        is_code_freq[code]["count"] += 1
        is_code_freq[code]["pages"].append(mat["Page Numbers"])

# 5. Build summary table
summary = []
for mat in materials_data:
    summary.append({
        "Material": mat["Material"],
        "IS Codes (Critical)": ", ".join(mat["Critical IS Codes"]) if mat["Critical IS Codes"] else ", ".join(mat["IS Codes"]),
        "Required Tests": "; ".join(mat["Required Tests"]),
        "Gradation/Mix Info": ", ".join(mat["Gradation/Mix Info"]),
        "Risk Notes": mat["Risk Notes"],
        "Page Numbers": mat["Page Numbers"]
    })

df = pd.DataFrame(summary)
df.to_csv('Industry_Intelligence_Summary.csv', index=False)
df.to_excel('Industry_Intelligence_Summary.xlsx', index=False)

# 6. Optional: JSON export for integration
with open('Industry_Intelligence_Summary.json', 'w') as f:
    json.dump(materials_data, f, indent=2)

# 7. IS Code frequency table
is_code_df = pd.DataFrame([
    {"IS Code": code, "Frequency": info["count"], "Pages": sorted(set(info["pages"]))}
    for code, info in is_code_freq.items()
])
is_code_df.to_csv('IS_Code_Frequencies.csv', index=False)

# 8. Display example
print("Industry-level summary table and JSON export complete!")
df.head(10)
```

```python
import pdfplumber, re, pandas as pd, json
from collections import defaultdict

# CONFIG: Set your PDF path here
pdf_path = '/content/drive/My Drive/Colab Notebooks/Sample_Contract.pdf'

# --- PARAMETERS FOR INDUSTRY-LEVEL EXTRACTION ---
IS_CODE_REGEX = r'IS\s*\d{3,5}(?:\s*[:\-]?\s*Part\s*\d+)?(?:\s*[:\-]?\s*\d{4})?'
SECTION_HEADER_REGEX = r'^\d+\.\d+(?:\.\d+)*[^\n]*'
RISKY_PHRASES = [
    "unless otherwise specified", "subject to engineer", "as directed", "unless approved",
    "to the satisfaction", "where applicable", "at the discretion"
]
CRITICAL_WORDS = ["shall", "must", "mandatory", "is to be", "no exception", "requirement", "will be"]
OPTIONAL_WORDS = ["may", "can", "should", "as desired", "recommended"]

# You may expand this list for your domain
KNOWN_MATERIALS = [
    "Cement", "Coarse Aggregate", "Fine Aggregate", "Chemical Admixtures", "Water", "Steel",
    "Batching", "Concrete", "Curing", "Formwork", "Reinforcement", "Grout"
]

# --- 1. Load and Chunk PDF ---
def chunk_page(text):
    # Split by double newlines or section headers (e.g., 4.1.1)
    # Returns a list of clause-like chunks
    if not text:
        return []
    # Add markers before section headers for splitting
    marked = re.sub(r'(^|\n)(\d+\.\d+(\.\d+)*[^\n]*)', r'\n\n\2', text)
    chunks = re.split(r'\n\n+', marked)
    return [c.strip() for c in chunks if c.strip()]

def get_sentences(text):
    # Simple sentence splitter
    return re.split(r'(?<=[.!?])\s+', text.strip())

# --- 2. Material and Clause Detection ---
def detect_material(chunk, material_names):
    # Look for material in heading or first 5 lines
    lines = chunk.split('\n')
    heading = lines[0] if lines else ""
    if any(material.lower() in heading.lower() for material in material_names):
        return True
    for line in lines[:5]:
        if any(material.lower() in line.lower() for material in material_names):
            return True
    return False

def get_material_from_heading(chunk, material_names):
    # Return the material found in heading, if any
    lines = chunk.split('\n')
    for line in lines[:2]:
        for material in material_names:
            if re.match(rf'(^|\s){re.escape(material)}(\s|$)', line, re.IGNORECASE):
                return material
    return None

# --- 3. Extraction Functions ---
def extract_is_codes(chunk):
    return list(set(re.findall(IS_CODE_REGEX, chunk, re.IGNORECASE)))

def extract_tables(chunk):
    # Only lines mentioning Table 4.x, mix, gradation, or proportion
    return [line for line in chunk.split('\n') if re.search(r'Table\s*\d+(\.\d+)?|mix|gradation|proportion', line, re.I)]

def extract_tests(chunk):
    # Return list of sentences mentioning test or compliance
    out = []
    for sent in get_sentences(chunk):
        if re.search(r'test|testing|conform|comply|strength|grade|sample|requirement', sent, re.IGNORECASE):
            tag = "Critical" if any(w in sent.lower() for w in CRITICAL_WORDS) else ("Optional" if any(w in sent.lower() for w in OPTIONAL_WORDS) else "")
            out.append(f"{sent.strip()} [{tag}]" if tag else sent.strip())
    return out

def extract_risks(chunk):
    return ", ".join({ph for ph in RISKY_PHRASES if ph in chunk.lower()})

def find_critical_is_codes(chunk, codes):
    # Only flag as "Critical" if the IS code and CRITICAL_WORD are in same sentence
    critical_codes = []
    for code in codes:
        for sent in get_sentences(chunk):
            if code in sent and any(w in sent.lower() for w in CRITICAL_WORDS):
                critical_codes.append(code)
    return list(set(critical_codes))

def get_definition(chunk):
    sentences = get_sentences(chunk)
    return " ".join(sentences[:2])  # Use 2 sentences for summary

def importance_score(appearances, critical_found, in_table):
    score = 0
    score += appearances * 2
    if critical_found:
        score += 10
    if in_table:
        score += 5
    return score

def get_clause_snippet(chunk, limit=350):
    return chunk[:limit] + ("..." if len(chunk) > limit else "")

# --- 4. MAIN EXTRACTION LOGIC ---
pages_text = []
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        pages_text.append({'page_num': i+1, 'text': text})

# Collect all results per material
material_summary = defaultdict(lambda: {
    "IS Codes": set(),
    "Critical IS Codes": set(),
    "Required Tests": set(),
    "Critical Tests": set(),
    "Gradation/Mix Info": set(),
    "Risk Notes": set(),
    "Definitions": [],
    "Page Numbers": set(),
    "Clause Snippets": [],
    "Appearances": 0,
    "In Table": False,
})

for p in pages_text:
    chunks = chunk_page(p['text'])
    for chunk in chunks:
        # Try to find a material heading or infer material from known list
        material = get_material_from_heading(chunk, KNOWN_MATERIALS)
        if not material:
            # Or if not in heading, check for any material in the first 5 lines
            for mat in KNOWN_MATERIALS:
                if detect_material(chunk, [mat]):
                    material = mat
                    break
        if material:
            mat = material_summary[material]
            mat["Appearances"] += 1
            codes = extract_is_codes(chunk)
            mat["IS Codes"].update(codes)
            critical_codes = find_critical_is_codes(chunk, codes)
            mat["Critical IS Codes"].update(critical_codes)
            tests = extract_tests(chunk)
            mat["Required Tests"].update(tests)
            mat["Critical Tests"].update([t for t in tests if "[Critical]" in t])
            tables = extract_tables(chunk)
            if tables: mat["In Table"] = True
            mat["Gradation/Mix Info"].update(tables)
            risks = extract_risks(chunk)
            if risks: mat["Risk Notes"].add(risks)
            # Only add new definitions if this is first chunk for material
            if mat["Appearances"] <= 2:
                mat["Definitions"].append(get_definition(chunk))
            mat["Page Numbers"].add(str(p['page_num']))
            mat["Clause Snippets"].append(get_clause_snippet(chunk))

# --- 5. Build Final Summary Table ---
rows = []
for material, info in material_summary.items():
    # Score for importance
    score = importance_score(
        info["Appearances"],
        bool(info["Critical IS Codes"] or info["Critical Tests"]),
        info["In Table"]
    )
    rows.append({
        "Material": material,
        "IS Codes (Critical)": ", ".join(sorted(info["Critical IS Codes"])) if info["Critical IS Codes"] else ", ".join(sorted(info["IS Codes"])),
        "All IS Codes": ", ".join(sorted(info["IS Codes"])),
        "Required Tests": "; ".join(sorted(info["Required Tests"])),
        "Critical Tests": "; ".join(sorted(info["Critical Tests"])),
        "Gradation/Mix Info": "; ".join(sorted(info["Gradation/Mix Info"])),
        "Definition": ". ".join(info["Definitions"]),
        "Risk Notes": "; ".join(sorted(info["Risk Notes"])),
        "Importance Score": score,
        "Pages": ", ".join(sorted(info["Page Numbers"])),
        "Clause Snippet": "\n---\n".join(info["Clause Snippets"][:2])  # Only first 2 for brevity
    })

df = pd.DataFrame(rows)
df.sort_values("Importance Score", ascending=False, inplace=True)
df.to_csv('Industry_Level_Material_Summary.csv', index=False)
df.to_excel('Industry_Level_Material_Summary.xlsx', index=False)

# JSON output for web integration
with open('Industry_Level_Material_Summary.json', 'w') as f:
    json.dump(rows, f, indent=2)

print("Industry-level extraction complete! Preview:")
print(df.head(10).to_string())
```

```python
import pdfplumber, re, pandas as pd
from collections import defaultdict

pdf_path = '/content/drive/My Drive/Colab Notebooks/Sample_Contract.pdf'

# Config: Known materials and phrase patterns
KNOWN_MATERIALS = [
    "Cement", "Coarse Aggregate", "Fine Aggregate", "Chemical Admixtures", "Water",
    "Steel", "Batching", "Concrete", "Curing", "Formwork", "Reinforcement", "Grout"
]
IS_CODE_REGEX = r'IS\s*\d{3,5}(?:\s*[:\-]?\s*Part\s*\d+)?(?:\s*[:\-]?\s*\d{4})?'
RISKY_PHRASES = [
    "unless otherwise specified", "subject to engineer", "as directed", "unless approved",
    "to the satisfaction", "where applicable", "at the discretion"
]
CRITICAL_WORDS = ["shall", "must", "mandatory", "requirement", "is to be", "no exception"]
OPTIONAL_WORDS = ["may", "can", "should", "as desired", "recommended"]
SECTION_HEADER_REGEX = r'^\d+\.\d+(\.\d+)*[^\n]*'

def chunk_page(text):
    # Split by section headers (e.g. 4.1.1), or double newlines
    if not text: return []
    marked = re.sub(r'(^|\n)(\d+\.\d+(\.\d+)*[^\n]*)', r'\n\n\2', text)
    chunks = re.split(r'\n\n+', marked)
    return [c.strip() for c in chunks if c.strip()]

def get_sentences(text):
    return re.split(r'(?<=[.!?])\s+', text.strip())

def get_heading(chunk):
    # Heading = first non-empty line
    lines = chunk.split('\n')
    for l in lines:
        if l.strip(): return l.strip()
    return ""

def get_material_name(heading, material_names):
    for mat in material_names:
        if re.search(rf'\b{mat}\b', heading, re.I): return mat
    return None

def extract_definition(chunk):
    # Use 2 sentences after the heading
    sents = get_sentences(chunk)
    return " ".join(sents[:2])

def extract_is_codes(chunk):
    return list(set(re.findall(IS_CODE_REGEX, chunk, re.IGNORECASE)))

def extract_critical_is_codes(chunk, codes):
    critical = set()
    for sent in get_sentences(chunk):
        for code in codes:
            if code in sent and any(w in sent.lower() for w in CRITICAL_WORDS):
                critical.add(code)
    return list(critical)

def extract_tests(chunk):
    out, out_critical = set(), set()
    for sent in get_sentences(chunk):
        if re.search(r'test|conform|comply|strength|grade|sample', sent, re.I):
            out.add(sent.strip())
            if any(w in sent.lower() for w in CRITICAL_WORDS):
                out_critical.add(sent.strip())
    return list(out), list(out_critical)

def extract_mix_info(chunk):
    lines = chunk.split('\n')
    return [l.strip() for l in lines if re.search(r'Table\s*\d+(\.\d+)?|mix|gradation|proportion', l, re.I)]

def extract_risks(chunk):
    return [ph for ph in RISKY_PHRASES if ph in chunk.lower()]

def get_clause_snippet(chunk, limit=280):
    sents = get_sentences(chunk)
    snippet = ""
    for s in sents:
        if len(snippet) + len(s) > limit: break
        snippet += s + " "
    return snippet.strip()

# 1. Read and chunk PDF by section
pages = []
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        for chunk in chunk_page(text):
            pages.append({'page': i+1, 'chunk': chunk})

# 2. Extract data per chunk and aggregate by material
mat_data = defaultdict(lambda: {
    "Definition": [],
    "IS Codes": set(),
    "Critical IS Codes": set(),
    "Required Tests": set(),
    "Critical Tests": set(),
    "Mix Info": set(),
    "Risk Notes": set(),
    "Pages": set(),
    "Clause Snippet": []
})

for item in pages:
    chunk, page = item['chunk'], item['page']
    heading = get_heading(chunk)
    mat = get_material_name(heading, KNOWN_MATERIALS)
    if not mat: continue  # Skip if not a material section

    mat_data[mat]["Pages"].add(str(page))
    definition = extract_definition(chunk)
    if definition: mat_data[mat]["Definition"].append(definition)
    codes = extract_is_codes(chunk)
    if codes: mat_data[mat]["IS Codes"].update(codes)
    crit_codes = extract_critical_is_codes(chunk, codes)
    if crit_codes: mat_data[mat]["Critical IS Codes"].update(crit_codes)
    tests, crit_tests = extract_tests(chunk)
    if tests: mat_data[mat]["Required Tests"].update(tests)
    if crit_tests: mat_data[mat]["Critical Tests"].update(crit_tests)
    mix = extract_mix_info(chunk)
    if mix: mat_data[mat]["Mix Info"].update(mix)
    risks = extract_risks(chunk)
    if risks: mat_data[mat]["Risk Notes"].update(risks)
    snippet = get_clause_snippet(chunk)
    if snippet: mat_data[mat]["Clause Snippet"].append(snippet)

# 3. Build clean, readable table
rows = []
for mat, info in mat_data.items():
    rows.append({
        "Material": mat,
        "Definition": " ".join(info["Definition"][:2]),
        "IS Codes": ", ".join(sorted(info["IS Codes"])),
        "Critical IS Codes": ", ".join(sorted(info["Critical IS Codes"])),
        "Required Tests": "; ".join(sorted(info["Required Tests"])),
        "Critical Tests": "; ".join(sorted(info["Critical Tests"])),
        "Mix Info": "; ".join(sorted(info["Mix Info"])),
        "Risk Notes": "; ".join(sorted(info["Risk Notes"])),
        "Pages": ", ".join(sorted(info["Pages"])),
        "Clause Snippet": " / ".join(info["Clause Snippet"][:2])
    })

df = pd.DataFrame(rows)
df.to_csv('Perfect_Material_Summary.csv', index=False)
df.to_excel('Perfect_Material_Summary.xlsx', index=False)
print("Done! Open the Excel/CSV for a clean, readable table.")
display(df)
```

```python
# HACKATHON COMPLIANT CONTRACT MATERIAL EXTRACTOR
# ------------------------------------------------
# This script extracts a strictly formatted table FIRST (4 main columns, concise, per guidelines)
# Then adds bonus "crazy" innovation columns for extra marks/insight.
# Output: Excel/CSV for checking, ready for PDF export.

import pdfplumber, re, pandas as pd
from collections import defaultdict

# ---- CONFIGURATION ----
pdf_path = '/content/drive/My Drive/Colab Notebooks/Sample_Contract.pdf'  # Change path as needed

# Known materials and relevant patterns
KNOWN_MATERIALS = [
    "Cement", "Coarse Aggregate", "Fine Aggregate", "Chemical Admixtures", "Water",
    "Steel", "Batching", "Concrete", "Curing", "Formwork", "Reinforcement", "Grout"
]
IS_CODE_REGEX = r'IS\s*\d{3,5}(?:\s*[:\-]?\s*Part\s*\d+)?(?:\s*[:\-]?\s*\d{4})?'
SECTION_HEADER_REGEX = r'^\d+\.\d+(\.\d+)*[^\n]*'
CRITICAL_WORDS = ["shall", "must", "mandatory", "requirement", "is to be", "no exception"]
RISKY_PHRASES = [
    "unless otherwise specified", "subject to engineer", "as directed", "unless approved",
    "to the satisfaction", "where applicable", "at the discretion"
]

# ---- BASIC TEXT UTILITIES ----
def chunk_page(text):
    # Split by section headers or double newlines (for technical docs)
    if not text: return []
    marked = re.sub(r'(^|\n)(\d+\.\d+(\.\d+)*[^\n]*)', r'\n\n\2', text)
    chunks = re.split(r'\n\n+', marked)
    return [c.strip() for c in chunks if c.strip()]

def get_sentences(text):
    return re.split(r'(?<=[.!?])\s+', text.strip())

def get_heading(chunk):
    lines = chunk.split('\n')
    for l in lines:
        if l.strip(): return l.strip()
    return ""

def get_material_name(heading, material_names):
    for mat in material_names:
        if re.search(rf'\b{mat}\b', heading, re.I): return mat
    return None

def extract_page_number(chunk, page):
    # Returns the page number if chunk contains data, else None
    return str(page)

# ---- 1. MAIN EXTRACTION LOGIC (STRICT 4-COLUMN FOCUS FIRST) ----
pages = []
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        for chunk in chunk_page(text):
            pages.append({'page': i+1, 'chunk': chunk})

# For each material, collect relevant info for the 4 main columns
main_table = []

for item in pages:
    chunk, page = item['chunk'], item['page']
    heading = get_heading(chunk)
    mat = get_material_name(heading, KNOWN_MATERIALS)
    if not mat: continue  # Only process known material sections

    # --- COLUMN 1: Material Name ---
    material_name = mat

    # --- COLUMN 2: Test Name/Reference Code/Standard (with page number) ---
    # Find all sentences with IS code or test keywords, add page number
    tests_refs = []
    for sent in get_sentences(chunk):
        codes = re.findall(IS_CODE_REGEX, sent)
        if codes or re.search(r'test|conform|comply|strength|grade|sample|requirement', sent, re.I):
            ref = ""
            if codes:
                ref = ", ".join(codes)
            test_name = ""
            # Try to extract test name (e.g., "Compressive strength test")
            m = re.search(r'([\w\s\-]+test)', sent, re.I)
            if m:
                test_name = m.group(1).strip()
            # Format: test name – IS code, Page X
            entry = ""
            if test_name:
                entry += test_name + " – "
            if ref:
                entry += ref
            if entry:
                entry += f", Page {page}"
                tests_refs.append(entry)
    if not tests_refs:
        tests_refs = ["No Information Available"]

    # --- COLUMN 3: Specific Material Type/Definition (short only) ---
    # Take only the first 1–2 lines after heading, max 100 chars
    def_lines = chunk.split('\n')[1:3]
    definition = ""
    for line in def_lines:
        if len(definition) + len(line) <= 100:
            definition += (line.strip() + " ")
    definition = definition.strip() if definition.strip() else "No Information Available"

    # --- COLUMN 4: Any Other Relevant Information (concise only) ---
    # Look for gradation table, silt content, etc. (max 2 items)
    other_info = []
    for line in chunk.split('\n'):
        if re.search(r'gradation|Table|silt content|proportion|fineness', line, re.I):
            # Try to keep only concise, non-redundant info
            if len(other_info) < 2 and line.strip() not in other_info:
                other_info.append(line.strip())
    if not other_info:
        other_info = ["No Information Available"]

    main_table.append({
        "Material Name": material_name,
        "Test Name/Reference Code/Standard (with page number)": "\n".join(tests_refs),
        "Specific Material Type/Material Definition": definition,
        "Any Other Relevant Information": "\n".join(other_info),
        # Below: fields for "crazy" ideas (added after core columns)
        "Risk Notes": "",
        "Criticality": "",
        "Clause Snippet": "",
        "Section/Clause Ref": "",
    })

# ---- 2. ADDING "CRAZY" INNOVATION COLUMNS ----
for i, row in enumerate(main_table):
    # ----- RISK NOTES -----
    mat = row["Material Name"]
    risks = []
    for item in pages:
        chunk, page = item['chunk'], item['page']
        heading = get_heading(chunk)
        mat2 = get_material_name(heading, KNOWN_MATERIALS)
        if mat2 == mat:
            for rp in RISKY_PHRASES:
                if rp in chunk.lower() and rp not in risks:
                    risks.append(rp)
    row["Risk Notes"] = "; ".join(risks) if risks else "No Information Available"

    # ----- CRITICALITY -----
    # Flag as "Critical" if any IS code/test sentence contains critical word
    is_critical = False
    for item in pages:
        chunk, page = item['chunk'], item['page']
        heading = get_heading(chunk)
        mat2 = get_material_name(heading, KNOWN_MATERIALS)
        if mat2 == mat:
            for sent in get_sentences(chunk):
                if any(w in sent.lower() for w in CRITICAL_WORDS) and (
                    re.search(IS_CODE_REGEX, sent) or re.search(r'test', sent, re.I)
                ):
                    is_critical = True
    row["Criticality"] = "Critical" if is_critical else "Normal"

    # ----- CLAUSE SNIPPET -----
    # Take first 1-2 sentences from the chunk as a snippet
    for item in pages:
        chunk, page = item['chunk'], item['page']
        heading = get_heading(chunk)
        mat2 = get_material_name(heading, KNOWN_MATERIALS)
        if mat2 == mat:
            sentences = get_sentences(chunk)
            snippet = " ".join(sentences[:2])[:200]
            if snippet and not row["Clause Snippet"]:
                row["Clause Snippet"] = snippet

    # ----- SECTION/CLAUSE REF -----
    # Try to find the section number in the heading
    for item in pages:
        chunk, page = item['chunk'], item['page']
        heading = get_heading(chunk)
        mat2 = get_material_name(heading, KNOWN_MATERIALS)
        if mat2 == mat:
            m = re.match(r'(\d+\.\d+(\.\d+)*)', heading)
            if m:
                row["Section/Clause Ref"] = m.group(1)
                break

# ---- 3. BUILD FINAL DATAFRAME AND EXPORT ----
df = pd.DataFrame(main_table)

# Strictly keep the 4 main columns first, then extras
ordered_cols = [
    "Material Name",
    "Test Name/Reference Code/Standard (with page number)",
    "Specific Material Type/Material Definition",
    "Any Other Relevant Information",
    "Risk Notes",
    "Criticality",
    "Clause Snippet",
    "Section/Clause Ref"
]
df = df[ordered_cols]

# Replace any empty or blank cell with "No Information Available"
df = df.fillna("No Information Available")
df.replace(r'^\s*$', "No Information Available", regex=True, inplace=True)

# Add Sl. No. as first column
df.insert(0, "Sl. No.", range(1, 1 + len(df)))

# Save to Excel/CSV for checking or PDF conversion
df.to_excel('Hackathon_Compliant_Material_Table.xlsx', index=False)
df.to_csv('Hackathon_Compliant_Material_Table.csv', index=False)

print("Extraction complete! Download the .xlsx for review.")
display(df.head(10))
```

```python
# HACKATHON 2025: COMPLIANT & AGGREGATING MATERIAL EXTRACTION SCRIPT
# ------------------------------------------------
# This script:
#   - Aggregates all info per material (so only ONE row per material!)
#   - Strictly follows the 4 main output columns as per guidelines
#   - Fills "No Information Available" only if truly nothing found
#   - Adds bonus innovation columns (Risk Notes, Criticality, Clause Snippet, Section Ref)
#   - Designed to work on ANY technical specification PDF
#   - Ready for Excel/CSV export and PDF table conversion

import pdfplumber, re, pandas as pd
from collections import defaultdict

# ---- CONFIGURATION ----
pdf_path = '/content/drive/My Drive/Colab Notebooks/Sample_Contract.pdf'  # Change as needed

KNOWN_MATERIALS = [
    "Cement", "Coarse Aggregate", "Fine Aggregate", "Chemical Admixtures", "Water",
    "Steel", "Batching", "Concrete", "Curing", "Formwork", "Reinforcement", "Grout"
]
IS_CODE_REGEX = r'IS\s*\d{3,5}(?:\s*[:\-]?\s*Part\s*\d+)?(?:\s*[:\-]?\s*\d{4})?'
TEST_KEYWORDS = r'test|conform|comply|strength|grade|sample|requirement|analysis|content'
CRITICAL_WORDS = ["shall", "must", "mandatory", "requirement", "is to be", "no exception"]
RISKY_PHRASES = [
    "unless otherwise specified", "subject to engineer", "as directed", "unless approved",
    "to the satisfaction", "where applicable", "at the discretion"
]

def chunk_page(text):
    if not text: return []
    marked = re.sub(r'(^|\n)(\d+\.\d+(\.\d+)*[^\n]*)', r'\n\n\2', text)
    chunks = re.split(r'\n\n+', marked)
    return [c.strip() for c in chunks if c.strip()]

def get_sentences(text):
    return re.split(r'(?<=[.!?])\s+', text.strip())

def get_heading(chunk):
    lines = chunk.split('\n')
    for l in lines:
        if l.strip(): return l.strip()
    return ""

def get_material_name(text, material_names):
    for mat in material_names:
        if re.search(rf'\b{mat}\b', text, re.I): return mat
    return None

# ---- 1. READ & CHUNK DOC ----
material_chunks = defaultdict(list)
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        for chunk in chunk_page(text):
            heading = get_heading(chunk)
            mat = get_material_name(heading, KNOWN_MATERIALS)
            if mat:
                material_chunks[mat].append({
                    'chunk': chunk,
                    'heading': heading,
                    'page': i+1
                })

# ---- 2. AGGREGATION BY MATERIAL ----
aggregated_rows = []
for idx, (mat, sections) in enumerate(material_chunks.items()):
    # --- COL 1: Material Name ---
    material_name = mat

    # --- COL 2: Test Name/Reference Code/Standard (with page number) ---
    test_refs = set()
    for sec in sections:
        for sent in get_sentences(sec['chunk']):
            codes = re.findall(IS_CODE_REGEX, sent)
            if codes or re.search(TEST_KEYWORDS, sent, re.I):
                test_name = ""
                m = re.search(r'([\w\s\-]+test)', sent, re.I)
                if m:
                    test_name = m.group(1).strip()
                ref = ", ".join(sorted(set(codes)))
                entry = ""
                if test_name:
                    entry += test_name + " – "
                if ref:
                    entry += ref
                if entry:
                    entry += f", Page {sec['page']}"
                    test_refs.add(entry)
    test_refs = sorted(test_refs)
    test_refs_str = "\n".join(test_refs) if test_refs else "No Information Available"

    # --- COL 3: Specific Material Type/Material Definition (short only) ---
    definitions = set()
    for sec in sections:
        lines = sec['chunk'].split('\n')[1:4]
        for line in lines:
            # Try to extract definitions/types (e.g. "43 Grade Ordinary Portland", "Aggregate most of which passes through 4.75 mm IS sieve")
            if len(line.strip()) > 3 and len(line.strip()) <= 120:
                # Heuristic: skip lines that are just headings or too long
                if re.search(r'grade|type|sieve|aggregate|concrete|admixture|steel|content|portland|mm|cement|water|fine|coarse', line, re.I):
                    definitions.add(line.strip())
    definition = "; ".join(sorted(definitions)) if definitions else "No Information Available"

    # --- COL 4: Any Other Relevant Information (concise only) ---
    other_info = set()
    for sec in sections:
        for line in sec['chunk'].split('\n'):
            if re.search(r'gradation|table|silt content|proportion|fineness|maximum|minimum|limit|not exceed', line, re.I):
                if len(line.strip()) > 3 and len(line.strip()) <= 120:
                    # Attach page number if not present
                    if f"Page {sec['page']}" not in line:
                        line = f"{line.strip()} (Page {sec['page']})"
                    other_info.add(line.strip())
    other_info_str = "; ".join(sorted(other_info)) if other_info else "No Information Available"

    # --- INNOVATION: Risk Notes ---
    risks = set()
    for sec in sections:
        for rp in RISKY_PHRASES:
            if rp in sec['chunk'].lower():
                risks.add(rp)
    risk_notes = "; ".join(sorted(risks)) if risks else "No Information Available"

    # --- INNOVATION: Criticality ---
    is_critical = False
    for sec in sections:
        for sent in get_sentences(sec['chunk']):
            if any(w in sent.lower() for w in CRITICAL_WORDS) and (
                re.search(IS_CODE_REGEX, sent) or re.search(TEST_KEYWORDS, sent, re.I)
            ):
                is_critical = True
    criticality = "Critical" if is_critical else "Normal"

    # --- INNOVATION: Clause Snippet ---
    snippet = ""
    for sec in sections:
        for sent in get_sentences(sec['chunk']):
            if len(snippet) < 200 and len(sent.strip()) > 10:
                snippet += sent.strip() + " "
            if len(snippet) > 180:
                break
        if snippet:
            break
    snippet = snippet.strip() if snippet else "No Information Available"

    # --- INNOVATION: Section/Clause Ref ---
    section_ref = "No Information Available"
    for sec in sections:
        m = re.match(r'(\d+\.\d+(\.\d+)*)', sec['heading'])
        if m:
            section_ref = m.group(1)
            break

    aggregated_rows.append({
        "Sl. No.": idx+1,
        "Material Name": material_name,
        "Test Name/Reference Code/Standard (with page number)": test_refs_str,
        "Specific Material Type/Material Definition": definition,
        "Any Other Relevant Information": other_info_str,
        # Innovation columns:
        "Risk Notes": risk_notes,
        "Criticality": criticality,
        "Clause Snippet": snippet,
        "Section/Clause Ref": section_ref
    })

# ---- 3. FINAL OUTPUT ----
ordered_cols = [
    "Sl. No.",
    "Material Name",
    "Test Name/Reference Code/Standard (with page number)",
    "Specific Material Type/Material Definition",
    "Any Other Relevant Information",
    "Risk Notes",
    "Criticality",
    "Clause Snippet",
    "Section/Clause Ref"
]
df = pd.DataFrame(aggregated_rows)[ordered_cols]

# Replace blanks with "No Information Available"
df = df.fillna("No Information Available")
df.replace(r'^\s*$', "No Information Available", regex=True, inplace=True)

# Save to Excel/CSV for review and PDF conversion
df.to_excel('Improved_Aggregated_Material_Table.xlsx', index=False)
df.to_csv('Improved_Aggregated_Material_Table.csv', index=False)

print("✅ Aggregation complete! One row per material, all info combined.")
print("Check Improved_Aggregated_Material_Table.xlsx for your hackathon-compliant output.")
display(df.head(10))
```

```python
import pdfplumber, re, pandas as pd
from collections import defaultdict

# Set your PDF path
pdf_path = '/content/Res-doc-2-21112016.pdf'

KNOWN_MATERIALS = [
    "Cement", "Coarse Aggregate", "Fine Aggregate", "Chemical Admixtures", "Water",
    "Steel", "Batching", "Concrete", "Curing", "Formwork", "Reinforcement", "Grout"
]
IS_CODE_REGEX = r'IS\s*\d{3,5}(?:\s*[:\-]?\s*Part\s*\d+)?(?:\s*[:\-]?\s*\d{4})?'
TEST_KEYWORDS = r'test|conform|comply|strength|grade|sample|requirement|analysis|content'
CRITICAL_WORDS = ["shall", "must", "mandatory", "requirement", "is to be", "no exception"]
RISKY_PHRASES = [
    "unless otherwise specified", "subject to engineer", "as directed", "unless approved",
    "to the satisfaction", "where applicable", "at the discretion"
]

def chunk_page(text):
    if not text: return []
    marked = re.sub(r'(^|\n)(\d+\.\d+(\.\d+)*[^\n]*)', r'\n\n\2', text)
    chunks = re.split(r'\n\n+', marked)
    return [c.strip() for c in chunks if c.strip()]

def get_sentences(text):
    return re.split(r'(?<=[.!?])\s+', text.strip())

def get_heading(chunk):
    lines = chunk.split('\n')
    for l in lines:
        if l.strip(): return l.strip()
    return ""

def get_material_name(text, material_names):
    for mat in material_names:
        if re.search(rf'\b{mat}\b', text, re.I): return mat
    return None

# ---- 1. READ & CHUNK DOC ----
material_chunks = defaultdict(list)
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        for chunk in chunk_page(text):
            heading = get_heading(chunk)
            mat = get_material_name(heading, KNOWN_MATERIALS)
            if mat:
                material_chunks[mat].append({
                    'chunk': chunk,
                    'heading': heading,
                    'page': i+1
                })

# ---- 2. AGGREGATION BY MATERIAL ----
aggregated_rows = []
for idx, (mat, sections) in enumerate(material_chunks.items()):
    # --- COL 1: Material Name ---
    material_name = mat

    # --- COL 2: Test Name/Reference Code/Standard (with page number) ---
    test_refs = set()
    for sec in sections:
        for sent in get_sentences(sec['chunk']):
            codes = re.findall(IS_CODE_REGEX, sent)
            if codes or re.search(TEST_KEYWORDS, sent, re.I):
                test_name = ""
                m = re.search(r'([\w\s\-]+test)', sent, re.I)
                if m:
                    test_name = m.group(1).strip()
                ref = ", ".join(sorted(set(codes)))
                entry = ""
                if test_name:
                    entry += test_name + " – "
                if ref:
                    entry += ref
                if entry:
                    entry += f", Page {sec['page']}"
                    test_refs.add(entry)
    test_refs = sorted(test_refs)
    test_refs_str = "\n".join(test_refs) if test_refs else "No Information Available"

    # --- COL 3: Specific Material Type/Material Definition (short only) ---
    definitions = set()
    for sec in sections:
        lines = sec['chunk'].split('\n')[1:4]
        for line in lines:
            if len(line.strip()) > 3 and len(line.strip()) <= 120:
                if re.search(r'grade|type|sieve|aggregate|concrete|admixture|steel|content|portland|mm|cement|water|fine|coarse', line, re.I):
                    definitions.add(line.strip())
    definition = "; ".join(sorted(definitions)) if definitions else "No Information Available"

    # --- COL 4: Any Other Relevant Information (concise only) ---
    other_info = set()
    for sec in sections:
        for line in sec['chunk'].split('\n'):
            if re.search(r'gradation|table|silt content|proportion|fineness|maximum|minimum|limit|not exceed', line, re.I):
                if len(line.strip()) > 3 and len(line.strip()) <= 120:
                    if f"Page {sec['page']}" not in line:
                        line = f"{line.strip()} (Page {sec['page']})"
                    other_info.add(line.strip())
    other_info_str = "; ".join(sorted(other_info)) if other_info else "No Information Available"

    # --- INNOVATION: Risk Notes ---
    risks = set()
    for sec in sections:
        for rp in RISKY_PHRASES:
            if rp in sec['chunk'].lower():
                risks.add(rp)
    risk_notes = "; ".join(sorted(risks)) if risks else "No Information Available"

    # --- INNOVATION: Criticality ---
    is_critical = False
    for sec in sections:
        for sent in get_sentences(sec['chunk']):
            if any(w in sent.lower() for w in CRITICAL_WORDS) and (
                re.search(IS_CODE_REGEX, sent) or re.search(TEST_KEYWORDS, sent, re.I)
            ):
                is_critical = True
    criticality = "Critical" if is_critical else "Normal"

    # --- INNOVATION: Clause Snippet ---
    snippet = ""
    for sec in sections:
        for sent in get_sentences(sec['chunk']):
            if len(snippet) < 200 and len(sent.strip()) > 10:
                snippet += sent.strip() + " "
            if len(snippet) > 180:
                break
        if snippet:
            break
    snippet = snippet.strip() if snippet else "No Information Available"

    # --- INNOVATION: Section/Clause Ref ---
    section_ref = "No Information Available"
    for sec in sections:
        m = re.match(r'(\d+\.\d+(\.\d+)*)', sec['heading'])
        if m:
            section_ref = m.group(1)
            break

    aggregated_rows.append({
        "Sl. No.": idx+1,
        "Material Name": material_name,
        "Test Name/Reference Code/Standard (with page number)": test_refs_str,
        "Specific Material Type/Material Definition": definition,
        "Any Other Relevant Information": other_info_str,
        "Risk Notes": risk_notes,
        "Criticality": criticality,
        "Clause Snippet": snippet,
        "Section/Clause Ref": section_ref
    })

ordered_cols = [
    "Sl. No.",
    "Material Name",
    "Test Name/Reference Code/Standard (with page number)",
    "Specific Material Type/Material Definition",
    "Any Other Relevant Information",
    "Risk Notes",
    "Criticality",
    "Clause Snippet",
    "Section/Clause Ref"
]
df = pd.DataFrame(aggregated_rows)[ordered_cols]
df = df.fillna("No Information Available")
df.replace(r'^\s*$', "No Information Available", regex=True, inplace=True)

# Save to Excel for download
df.to_excel('Improved_Aggregated_Material_Table.xlsx', index=False)
print("✅ Extraction complete! Downloading Excel file...")

# Download the Excel file in Colab
from google.colab import files
files.download('Improved_Aggregated_Material_Table.xlsx')
```
